```
install.packages("ergm")
library(ergm)
```

```

members <- read.table("/content/ownerPoli.txt",header=FALSE,sep=" ")
attributes <- read.table("/content/ownerPoliAttr.txt",header=FALSE)
L <- readLines("/content/ownerPoliLabels.txt")
DF <- read.table(text = L, sep = "0",header=FALSE, stringsAsFactors=F, strip.white=T, as.is=T)
attributeMatrix <- cbind(DF,attributes)
politicalNet <- network(members,vertex.attr=attributeMatrix,vertex.attrnames = c("label","political"),directed=F, hyper=F, loops=F, multiple=F, bipartite=F)
plot(politicalNet,label=get.vertex.attribute(politicalNet, "label"), vertex.col=c("blue","red")[1+(get.vertex.attribute(politicalNet, "political")==0)])
```

```
attrErgm <- ergm(politicalNet~edges+nodefactor("political")+nodematch("political")) #Estimate the model
summary(attrErgm) #Summarize the model
```

```
plot(simulate(attrErgm),
          vertex.col=c("blue","red")[ 1+(get.vertex.attribute(politicalNet, "political")==0)])
```

```
ownerNet <- gof(attrErgm)
```

```
plot(ownerNet)
```

```
members <- read.table("/content/resObsAnalysis.txt",header=FALSE,sep=" ")
edgeCovariates <- as.matrix(read.table("/content/edgeCovData.txt",header=FALSE,sep=" "))
resObsNet <- network(members,directed=F, hyper=F, loops=F, multiple=F, bipartite=11)
plot(resObsNet)
```

```
edgeCovariates
```

```
resObsErgm <- ergm(resObsNet~edges+edgecov(edgeCovariates),bipartite=11)
summary(resObsErgm)
```

```
plot(simulate(resObsErgm))
```
