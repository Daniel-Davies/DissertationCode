```python
def timeRestrictedBusinessGraph(upToTime):
    G = nx.Graph()
    
    data = customByBusiness()
    incorpDate = incorporationDates()

    names = data.keys()
    for k,i in enumerate(names): 
        if incorpDate[i] <= upToTime:
            G.add_node(i)
    
    
    for name1 in names:
        connects = data[name1]
        for item in connects:
            if incorpDate[item] <= upToTime and incorpDate[name1] <= upToTime:
                G.add_edge(name1,item)
    
    labels={}
    for k,i in enumerate(names): 
        if incorpDate[i] <= upToTime:
            labels[i] = i
    return G, labels

def plotGraph(G,labels,colors=None):
    pos=nx.spring_layout(G,k=1)
    nx.draw(G,pos,node_color=colors)
    for label in pos:
        x,y = pos[label]
        plt.text(x,y+0.035,s=labels[label],horizontalalignment='center')

        
for item in range(1996,2020):
    G,labels = timeRestrictedBusinessGraph(1996)
    file_ = nx.to_numpy_matrix(G)
    np.savetxt(str(item)+".txt",file_, "%d") 
    
# plotGraph(G,labels)
```
