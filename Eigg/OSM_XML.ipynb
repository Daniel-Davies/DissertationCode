```python
import untangle
import requests
import xmltodict
import pickle
```

```python
xml = requests.get('https://api.openstreetmap.org/api/0.6/map?bbox=-6.2259927,56.8575519,-6.0930367,56.9482939')
xml = xml.text
obj = xmltodict.parse(xml)
```

```python
from collections import defaultdict

nodes = obj['osm']['node']
#print(len(nodes))

dict_ = defaultdict(list)

mainObj = obj['osm']['node']
for i in range(len(nodes)):
    if 'tag' in mainObj[i].keys():
        for k in range(len(mainObj[i]['tag'])):
            try:
                tg = mainObj[i]['tag'][k]['@k']
                if tg not in ['ref','prominence','source:name','created_by', 'source', 'uk_postcode_centroid', 'wikipedia', 'wikidata', 'generator:source', 'generator:type'] and 'seamark' not in tg:
                    dict_[(mainObj[i]['tag'][k]['@k'])].append((mainObj[i]['tag'][k]['@v'],[mainObj[i]['@lat'], mainObj[i]['@lon']] ))
            except Exception as e: pass
```

```python
dict_
```

```python
kz = dict_.keys()
```

```python
for k in kz:
    items = dict_[k]
    for item in items:
        name, (lat,ln) = item
        print(k + "--" + name + "--" + lat + "--" + ln)
```
