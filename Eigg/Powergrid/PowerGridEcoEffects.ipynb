```python
def expand(tup):
    lst = []
    year, list_ = tup
    for item in list_:
        lst.append((year,item))
    return lst
```

```python
def makeBoxPlot(toExpand):
    labels, dat = [*zip(*toExpand.items())]  # 'transpose' items to parallel key, value lists
    labels = labels[::-1]
    dat = dat[::-1]
    sns.boxplot(data = dat)
    plt.xticks(range(0, len(labels)), labels)
    plt.show()
```
